```python
from mxnet import gluon
from mxnet import nd
from mxnet import autograd as ag
```

```python
num_inputs = 2
num_examples = 1000
true_w = [2, -3.4]
true_b = 4.2
X = nd.random_normal(shape=(num_examples, num_inputs))
y = X[:, 0] * true_w[0] + X[:, 1] * true_w[1] + true_b
y = 0.01 * nd.random_normal(shape=y.shape) + y
```

```python
y.shape
```

```python
import random
def data_iter(X, y, batch_size):
    index = nd.arange(len(X))
    random.shuffle(index)
    for i in range(0,num_examples, batch_size):
        j = nd.array(index[i:min(i+batch_size, num_examples)])
        yield X.take(j), y.take(j)
```

```python
def net(X):
    return nd.dot(X, w) + b
```

```python
def square_loss(y_hat, y):
    return 0.5 *nd.square(y_hat-y.reshape(y_hat.shape))
```

```python
def sgd(params, lr):
    for param in params:
        param[:] = param - lr * param.grad
```

```python
w = nd.random_normal(shape=(num_inputs,1))
b = nd.zeros(1)
params = [w, b]
for param in params:
    param.attach_grad()
```

```python
num_epochs = 3
batch_size = 100
lr = 0.01
for epoch in range(num_epochs):
    total_loss = 0.
    for data, label in data_iter(X,y,batch_size):
        with ag.record():
            output = net(data)
            loss = square_loss(output, label)
        loss.backward()
        sgd(params, lr)
        total_loss += loss.sum().asscalar()
    print("Epoch: %d, loss: %f" %(epoch, total_loss/num_examples))
```

```python
datasets = gluon.data.ArrayDataset(X,y)
data_loder = gluon.data.DataLoader(datasets, batch_size, shuffle = True)
```

```python
net = gluon.nn.Sequential()
net.add(gluon.nn.Dense(1))
net.initialize()
```

```python
square_loss = gluon.loss.L2Loss()
trainer = gluon.Trainer(net.collect_params(), 'sgd', {'learning_rate':lr})
```

```python
num_epochs = 3
batch_size = 10
lr = 0.1
for epoch in range(num_epochs):
    total_loss = 0.
    for data, label in data_loder:
        with ag.record():
            output = net(data)
            loss = square_loss(output, label)
        loss.backward()
        trainer.step(batch_size)
        total_loss += loss.sum().asscalar()
    print("Epoch: %d, loss: %f" %(epoch, total_loss/num_examples))
```

```python
x = nd.array([3])
x = x.tile(reps=(1000,1))
w = nd.ones(1)
y =  2 * x + 1
```

```python
w.attach_grad()
batch_size = 10
for i in range(1):
    for data, label in data_iter(x,y,batch_size):
        with ag.record():
            output = w * data + 1
        output.backward()
        print(w.grad)
```
